```python
import openpyxl
import pandas as pd
import googletrans
from googletrans import Translator
```

```python
xl_file2 = pd.read_excel('Downloads/BDO_FPCGZ.xlsx', encoding='utf-8')
```

```python
xl_file2
```

```python
translator = Translator()

# make a deep copy of the data frame
df_en1 = xl_file2.copy()

# translate columns' name using rename function
df_en1.rename(columns=lambda x: translator.translate(x, src= 'pt',dest='en').text, inplace=True)

# translated column names
df_en1.columns
```

```python
df_en1
```

```python
df_en1 = df_en1.dropna(axis = 1, how = 'all')
df_en1
```

```python
df_en1 = df_en1.fillna('none')
```

```python
df_en1
```

```python
translations = {}
for column in df_en1.columns:
    # unique elements of the column
    unique_elements = df_en1[column].iloc[0:2].unique()
    for elements in unique_elements:
        translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
    
print(translations)
```

```python
df_en1.replace(translations, inplace = True)
df_en1
```

```python
writer = pd.ExcelWriter('Downloads/new_excel_brazil.xlsx', engine='xlsxwriter')

df_en1.to_excel(writer, index=False)
workbook  = writer.book
worksheet = writer.sheets['Sheet1']

#merge_format = workbook.add_format({'align': 'center'})
formatted = workbook.add_format({
'bold': True,
'border': 1,
'align': 'center',
'valign': 'vcenter',
})
formatted1 = workbook.add_format({
'border': 1,
'align': 'center',
'valign': 'vcenter',
})

#worksheet.conditional_format( 'A1:S3' , { 'type' : 'no_blanks','format' : formatted} )
worksheet.conditional_format( 'B1:S3' , { 'type' : 'no_blanks','format' : formatted} )
worksheet.conditional_format( 'A4:S12' , { 'type' : 'no_blanks','format' : formatted1} )
trial = worksheet.conditional_format( 'A1:A3' , { 'type' : 'no_blanks','format' : formatted} )

worksheet.merge_range('A1:A3', trial)


writer.save()
```
