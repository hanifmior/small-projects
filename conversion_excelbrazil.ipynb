```python
import openpyxl
import pandas as pd
```

```python
import_file2 = pd.read_excel('Downloads/EXCELBRAZIL.xlsx', encoding='utf-8')
```

```python
import_file2
```

```python
df = import_file2.copy()
outputpsi1 = {}
outputpsi2 = {}
outputpsi3 = {}
outputpsi4 = {}
for column in df.columns:
    # unique elements of the column
    unique_elements1 = df['PDG'].iloc[2:11].unique()
    unique_elements2 = df['Well Head Pressure & temperature '].iloc[2:11].unique()
    #unique_elements1 = df['Well Head Pressure & temperature'].iloc[2:10].unique()
    unique_elements3 = df['Line Pressure'].iloc[2:11].unique()
    unique_elements4 = df['Unnamed: 16'].iloc[2:11].unique()
    for elements in unique_elements1:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputpsi1[elements] = elements * (98.0665 / 6.89475729)
    for elements in unique_elements2:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputpsi2[elements] = elements * (98.0665 / 6.89475729)
    for elements in unique_elements3:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputpsi3[elements] = elements * (98.0665 / 6.89475729)
    for elements in unique_elements4:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputpsi4[elements] = elements * (98.0665 / 6.89475729)
        
print(outputpsi1)
print(outputpsi2)
print(outputpsi3)
print(outputpsi4)
```

```python
df.replace(outputpsi1, inplace = True)
df.replace(outputpsi2, inplace = True)
df.replace(outputpsi3, inplace = True)
df.replace(outputpsi4, inplace = True)
df
```

```python
df.replace('kg / cm2', 'psig',inplace=True)
df
```

```python
outputbbl1 = {}
outputbbl2 = {}
outputbbl3 = {}
outputbbl4 = {}
outputscf5 = {}
for column in df.columns:
    # unique elements of the column
    unique_elementsA = df['Gross Production'].iloc[2:11].unique()
    unique_elementsB = df['Produced Oil'].iloc[2:11].unique()
    unique_elementsC = df['Potential Oil'].iloc[2:11].unique()
    unique_elementsD = df['Loses'].iloc[2:11].unique()
    unique_elementsE = df['Gas Lift'].iloc[2:11].unique()
    for elements in unique_elementsA:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputbbl1[elements] = elements * 6.29
    for elements in unique_elementsB:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputbbl2[elements] = elements * 6.29
    for elements in unique_elementsC:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputbbl3[elements] = elements * 6.29
    for elements in unique_elementsD:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputbbl4[elements] = elements * 6.29
    for elements in unique_elementsE:
       #translations[elements]= translator.translate(elements,src = 'pt',dest='en').text 
       outputscf5[elements] = elements * 35.3146667214886
        
print(outputbbl1)
print(outputbbl2)
print(outputbbl3)
print(outputbbl4)
print(outputscf5)
```

```python
df.replace(outputbbl1, inplace = True)
df.replace(outputbbl2, inplace = True)
df.replace(outputbbl3, inplace = True)
df.replace(outputbbl4, inplace = True)
df.replace(outputscf5, inplace = True)
df
```
